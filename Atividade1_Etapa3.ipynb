O refino de óleo de soja em Campo Mourão(PR) conta com uma produção diária de 720 tonelada em 2021 (Considerar uma variação diaria de +-50 toneladas)

```python
import numpy as np
import pandas as pd
#Definir o intevalo de tempo para analise
dtime = np.arange(np.datetime64("2021-01-01"), np.datetime64("2022-01-01"), np.timedelta64(1, "D"))
df = pd.DataFrame()
df['Dia'] = dtime
#Gerar numeros aleatórios(Randon)
df['Toneladas de Oleo de Soja'] = np.random.uniform(770, 670, size=df.shape[0]) #Intervalos de Valores
df = df.set_index(df['Dia'])
df['Mês'] = df.index.month
print(df)
df['Toneladas de Oleo de Soja'].plot(color='blue')
```
